```python
data['Date']=pd.to_datetime(data.Date)
```

```python
data.set_index('Date')
```

```python
dataset=data.groupby(['Date'])['Cured','Deaths','Confirmed'].sum().reset_index()
```

```python
import fbprophet
from fbprophet import Prophet
```

```python
dataset.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(100,50))
dataset['Confirmed'].plot(kind='bar')

```

```python
df=dataset.rename(columns={'Date':'ds','Confirmed':'y'})
```

```python
m=Prophet()
```

```python
model=m.fit(df)
```

```python
model.seasonalities
```

```python
future=model.make_future_dataframe(periods=30,freq='D')
```

```python
future.tail()
```

```python
prediction=model.predict(future)
```

```python
prediction
```

```python
prediction[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
pd.options.display.float_format = '{:.5f}'.format
prediction
```

```python
model.plot(prediction,)
```
